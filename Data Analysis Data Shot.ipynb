### DATA VISUALIZATION
#### UNIVARIATE ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample dataframe
data = {
    'Age': np.random.randint(18, 70, 1000),
    'Height': np.random.normal(160, 10, 1000),
    'Weight': np.random.normal(60, 5, 1000),
    'Gender': np.random.choice(['Male', 'Female'], 1000),
    'Education': np.random.choice(['High School', 'College', 'Graduate'], 1000)
}

df = pd.DataFrame(data)

## 1. Categorical Data for Univariate
# Plot histograms for numerical variables
numerical_cols = df.select_dtypes(include=['int', 'float']).columns
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Plot Dist plots for numerical variables
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.distplot(df[col])    
    plt.title(f'Dist Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()    
    
# Plot KDE plots for numerical variables
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(df[col], shade=True, color='skyblue')
    plt.title(f'KDE Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()

# Plot BOX plots for numerical variables
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(df[col])
    plt.title(f'BOX Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

```python
# 2. Numerical Data for Univariate
# Plot countplot plots for categorical variables
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    df[col].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Bar Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Plot bar plots for categorical variables
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=col)

    plt.title(f'countplot of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
    
    
# Plot pie charts for categorical variables
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    df[col].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral'])
    plt.title(f'Pie Chart of {col}')
    plt.ylabel('')
    plt.show()
```

## Multivariate 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample dataframe
data = {
    'Age': np.random.randint(18, 70, 1000),
    'Height': np.random.normal(160, 10, 1000),
    'Weight': np.random.normal(60, 5, 1000),
    'Gender': np.random.choice(['Male', 'Female'], 1000),
    'Education': np.random.choice(['High School', 'College', 'Graduate'], 1000)
}
df = pd.DataFrame(data)

# Scatterplot (Numerical - Numerical)
sns.scatterplot(data=df, x='Age', y='Height')
plt.title('Scatterplot (Age vs Height)')
plt.show()

# Bar Plot (Numerical - Categorical)
sns.barplot(data=df, x='Gender', y='Age')
plt.title('Bar Plot (Gender vs Age)')
plt.show()

# Box Plot (Numerical - Categorical)
sns.boxplot(data=df, x='Education', y='Age')
plt.title('Box Plot (Education vs Age)')
plt.show()

# Distplot (Numerical - Categorical)
sns.distplot(df['Age'][df['Gender'] == 'Male'], hist=False, label='Male')
sns.distplot(df['Age'][df['Gender'] == 'Female'], hist=False, label='Female')
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distplot (Age by Gender)')
plt.legend()
plt.show()

# HeatMap (Categorical - Categorical)
plt.figure(figsize=(8, 6))
pd.crosstab(df['Education'], df['Gender'])
sns.heatmap(pd.crosstab(df['Education'], df['Gender']), cmap='coolwarm', annot=True, fmt='d')
plt.title('HeatMap (Education vs Gender)')
plt.show()


# Pairplot
sns.pairplot(df[['Age', 'Height', 'Weight']])
plt.title('Pairplot')
plt.show()

# Lineplot (Numerical - Numerical)
sns.lineplot(data=df, x='Age', y='Height')
plt.title('Lineplot (Age vs Height)')
plt.show()
```

```python
import pandas as pd

# Sample DataFrame with a date column
data = {'date_column': ['2023-01-01', '2023-02-15', '2023-03-20','2023-05-31']}
df = pd.DataFrame(data)

# Convert date column to datetime type
df['date_column'] = pd.to_datetime(df['date_column'])

# Extracting features
df['year'] = df['date_column'].dt.year
df['month'] = df['date_column'].dt.month
df['day'] = df['date_column'].dt.day
df['weekday'] = df['date_column'].dt.weekday  # Monday is 0, Sunday is 6
df['is_weekend'] = df['date_column'].dt.weekday >= 5  # True if weekend, False otherwise
df['quarter'] = df['date_column'].dt.quarter
df['week_of_year'] = df['date_column'].dt.isocalendar().week
df['day_of_year'] = df['date_column'].dt.dayofyear
df['is_leap_year'] = df['date_column'].dt.is_leap_year

# Extracting additional features

df['week'] = df['date_column'].dt.week  # Week number of the year
df['day_of_week'] = df['date_column'].dt.dayofweek  # Day of the week (Monday=0, Sunday=6)
df['weekend'] = df['date_column'].dt.dayofweek.isin([5, 6])  # True if weekend, False otherwise
df['is_month_start'] = df['date_column'].dt.is_month_start
df['is_month_end'] = df['date_column'].dt.is_month_end
df['quarter_start'] = df['date_column'].dt.is_quarter_start
df['quarter_end'] = df['date_column'].dt.is_quarter_end
df['hour'] = df['date_column'].dt.hour
df['minute'] = df['date_column'].dt.minute
df['second'] = df['date_column'].dt.second


# Print DataFrame with new features
print(df)
```

```python
import datetime

# Creating datetime objects
now = datetime.datetime.now()
future_date = datetime.datetime(2024, 12, 31, 23, 59, 59)
past_date = datetime.datetime(2023, 1, 1)

# Formatting datetime objects
formatted_now = now.strftime("%Y-%m-%d %H:%M:%S")
parsed_date = datetime.datetime.strptime("2023-05-15", "%Y-%m-%d")

# Accessing attributes
year_now = now.year
month_now = now.month
day_now = now.day

# Arithmetic operations
difference = future_date - past_date
future_delta = datetime.timedelta(days=365)
new_date = past_date + future_delta

# Comparison
is_future = future_date > now
is_past = past_date < now

# Timezone conversion
utc_now = now.astimezone(datetime.timezone.utc)

# Extracting information
weekday_now = now.weekday()  # Monday is 0
iso_weekday_now = now.isoweekday()  # Monday is 1
time_tuple_now = now.timetuple()

# Printing results
print("Current datetime:", now)
print("Future date:", future_date)
print("Past date:", past_date)
print("Formatted datetime:", formatted_now)
print("Parsed datetime:", parsed_date)
print("Year:", year_now)
print("Month:", month_now)
print("Day:", day_now)
print("Difference:", difference)
print("Future delta:", future_delta)
print("New date:", new_date)
print("Is future date:", is_future)
print("Is past date:", is_past)
print("UTC now:", utc_now)
print("Weekday:", weekday_now)
print("ISO weekday:", iso_weekday_now)
print("Time tuple:", time_tuple_now)
```
